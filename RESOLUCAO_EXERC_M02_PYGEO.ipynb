E2 - Criando pontos e plotando em um mapa

Nosso objetivo é plotar uma série de pares de coordenadas que serão lidas do arquivo csv escooter.csv. Esse arquivo possui 1008 linhas que representam a localização e horário dos pontos de destino e origem de viagens patinetes elétricos em Tempe - AZ, EUA. Esses dados são amostras de alguns dados que uso em minha pesquisa. Você encontrará as seguintes colunas nessa tabela de dados:

Timestap_Origin: Horário na localização de origem
Scooter_ID: número identificador do patinete elétrico
Long_Origin: Longitude de origem
Lat_Origin: Latitude de origem
Battery_Origin: Nível da bateria na origem
Timestamp_Destination: Horário na localização de destino
Long_Destination:Longitude do destino
Lat_Destination:Latitude de destino
Battery_Destination: Nível da bateria no destino

a) Leia os dados com a biblioteca pandas

```python
import pandas as pd

txt_path='/content/drive/MyDrive/Colab Notebooks/MATERIAL_CURSO_DRIVE/escooter.csv'
dados=pd.read_csv(txt_path)

dados.head()
```

b) Crie uma coluna vazia chamada geometry

```python
dados['geometry']=None
dados.head()
```

c) Itere sob o dataframe e insira objetos do tipo LineString que conectem a origem e o destino de cada registro

```python
from shapely.geometry import LineString
```

```python
for index, row in dados.iterrows():
  dados.loc[index, 'geometry']= LineString([(row['Long_Origin'], row['Lat_Origin']), (row['Long_Destination'], row ['Lat_Destination'])])
```

```python
dados.head()
```

d) Converta para um GeoDataFrame e defina o sistema de coordenadas LATLONG WGS84

```python
type(dados)
```

```python
gdf=gpd.GeoDataFrame(dados)
```

```python
gdf.head()
```

```python
type(gdf)
```

```python
print(gdf.crs)
```

```python
# Definindo um sistema de coordenadas
gdf=gdf.set_crs(crs='epsg:4326')
gdf.crs
```

e) Salve o geodataframe como shapefile

```python
gdf.to_file('escooter_WGS84_.shp')
```

f) Crie uma figura dos dados com um mapa de fundo

E3 - Cálculo de distâncias

Nosso objetivo é calcular a distância em metros que cada patinete elétrico percorreu e responder as seguintes perguntas:

Qual foi a menor distância percorrida em km?
Qual foi a distância média percorrida em km?
Qual foi a distância máxima percorrida em km?
Reprojete os dados da projeção WGS84 para o EPSG:32612 (UTM Zona 12N - Zona UTM para o AZ - EUA) para transformar os dados para um sistema projetado (métrico).

```python
# Reprojetar para o sistema UTM ZONA 12N (ATENDE AO ARIZONA EUA)

gdf_utm=gdf.to_crs('epsg:32612')
gdf_utm.crs
```

```python
gdf_utm['dist']= None

```

```python
gdf_utm.head()
```

```python
for index, row in gdf_utm.iterrows():
  gdf_utm.loc[index, 'dist']=row['geometry'].length/1000
```

```python
gdf_utm.head()
```

```python
# Verificar valor minimo
min(gdf_utm['dist'])
```

```python
# Verificar valor máximo
max(gdf_utm['dist'])
```

```python
# Verificar o valor médio da numpy
import numpy as np
np.mean(gdf_utm['dist'])
```
